```python
def metadata(sample):
    sampleID = int(sample.split("_")[1])
    return sampleID
```

```python
sampleID = metadata("BB_941")
```

```python
age = biodata_df2[sampleID][3]
bbtype = biodata_df2[sampleID][5]
ethnicity = biodata_df2[sampleID][1]
gender = biodata_df2[sampleID][2]
location = biodata_df2[sampleID][6]
sampleid = sampleID
```

```python
metadata_dict = {        
        "AGE": age,
        "BBTYPE": bbtype,
        "ETHNICITY": ethnicity,
        "GENDER": gender,
        "LOCATION": location,
        "SAMPLEID": sampleid
}
```

```python
def samplevalue(sample):
    chosen = biodata_df.loc[:,("OTU ID #", sample, 'Lowest Taxonomic Level of Bacteria/Archaea Found')]
    chosen[sample] = pd.to_numeric(chosen[sample], downcast='integer', errors='coerce')
    chosen = chosen.sort_values(by=sample, ascending=False)
    otu_descriptions = chosen['Lowest Taxonomic Level of Bacteria/Archaea Found'].tolist()
    otu_ids = chosen['OTU ID #'].tolist()
    sample_values = chosen[sample].tolist()
    samples = [{"otu_ids": otu_ids}, {"sample_values": sample_values}]
    return samples
```
